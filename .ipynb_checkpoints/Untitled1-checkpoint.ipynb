```python
%matplotlib inline
import warnings

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import torch
from torch import optim
from torch.nn import functional as F
from sklearn.preprocessing import LabelEncoder
from torch.utils.data import Dataset, DataLoader
from nbeats_pytorch.model import NBeatsNet

warnings.filterwarnings(action='ignore', message='Setting attributes')
```

```python
# Load data
df = pd.read_csv("Data\MPS LSOA Level Crime (most recent 24 months).csv")
df.head().T
```

```python
# Unpivot the DataFrame
df = df.melt(id_vars=["LSOA Code", "LSOA Name", "Borough", "Major Category", "Minor Category"],
             var_name="date", value_name="frequency")

# Convert the date column to a datetime and set it as the index
df["date"] = pd.to_datetime(df["date"], format="%Y%m")
df.set_index("date", inplace=True)
```

```python
df.head()
```

```python
# drop the LSOA code column since its the same with LSOA name
df.drop('LSOA Code', inplace = True, axis = 1)
```

```python
# Encode categorical columns

encoder = LabelEncoder()
df['LSOA Name'] = encoder.fit_transform(df['LSOA Name'])
df['Borough'] = encoder.fit_transform(df['Borough'])
df['Major Category'] = encoder.fit_transform(df['Major Category'])
df['Minor Category'] = encoder.fit_transform(df['Minor Category'])
```

```python
df
```

```python
type(df)
```

```python
forecast_length = 1
backcast_length = 24

df = df.values.flatten()  # just keep np array here for simplicity.

# data backcast/forecast generation.
x, y = [], []
for epoch in range(backcast_length, len(df) - forecast_length):
    x.append(df[epoch - backcast_length:epoch])
    y.append(df[epoch:epoch + forecast_length])
x = np.array(x)
y = np.array(y)
```

```python
x.shape
```

```python
y.shape
```

```python
# Define a custom PyTorch dataset for the data
class CustomDataset(Dataset):
    def __init__(self, df):
        self.df = df

    def __len__(self):
        return len(self.df)

    def __getitem__(self, index):
        row = self.df.iloc[index]
        x = torch.FloatTensor(row[:-1])
        y = torch.FloatTensor([row[-1]])
        return x, y
```

```python
# Split the data into train and valid sets with an 80:20 ratio
train_size = int(len(df) * 0.8)
x_train, y_train = x[:train_size], y[:train_size]
x_test, y_test = x[train_size:], y[train_size:]
```

```python
# normalization.
norm_constant = np.max(x_train)
x_train, y_train = x_train / norm_constant, y_train / norm_constant
x_test, y_test = x_test / norm_constant, y_test / norm_constant
```

```python
# plot utils.
def plot_scatter(*args, **kwargs):
    plt.plot(*args, **kwargs)
    plt.scatter(*args, **kwargs)


# simple batcher.
def data_generator(x, y, size):
    assert len(x) == len(y)
    batches = []
    for ii in range(0, len(x), size):
        batches.append((x[ii:ii + size], y[ii:ii + size]))
    for batch in batches:
        yield batch
```

```python
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Define the N-BEATS model
model = NBeatsNet(
    stack_types=[NBeatsNet.GENERIC_BLOCK, NBeatsNet.GENERIC_BLOCK],
    forecast_length=1,
    thetas_dim=[8, 8],
    nb_blocks_per_stack=3,
    backcast_length=24,
    hidden_layer_units=24,
    share_weights_in_stack=False
)
optimiser = optim.Adam(lr=1e-4, params=model.parameters())
```

```python
grad_step = 0
batch_size = 12
for epoch in range(1000):
    # train.
    model.train()
    train_loss = []
    for x_train_batch, y_train_batch in data_generator(x_train, y_train, batch_size):
        grad_step += 1
        optimiser.zero_grad()
        _, forecast = model(torch.tensor(x_train_batch, dtype=torch.float).to(model.device))
        loss = F.mse_loss(forecast, torch.tensor(y_train_batch, dtype=torch.float).to(model.device))
        train_loss.append(loss.item())
        loss.backward()
        optimiser.step()
    train_loss = np.mean(train_loss)

    # test.
    model.eval()
    _, forecast = model(torch.tensor(x_test, dtype=torch.float))
    test_loss = F.mse_loss(forecast, torch.tensor(y_test, dtype=torch.float)).item()
    p = forecast.detach().numpy()
    if epoch % 100 == 0:
        subplots = [221, 222, 223, 224]
        plt.figure(1)
        for plot_id, i in enumerate(np.random.choice(range(len(x_test)), size=4, replace=False)):
            ff, xx, yy = p[i] * norm_constant, x_test[i] * norm_constant, y_test[i] * norm_constant
            plt.subplot(subplots[plot_id])
            plt.grid()
            plot_scatter(range(0, backcast_length), xx, color='b')
            plot_scatter(range(backcast_length, backcast_length + forecast_length), yy, color='g')
            plot_scatter(range(backcast_length, backcast_length + forecast_length), ff, color='r')
        plt.show()

        with torch.no_grad():
            save(model, optimiser, grad_step)
        print(f'epoch = {str(epoch).zfill(4)}, '
              f'grad_step = {str(grad_step).zfill(6)}, '
              f'tr_loss (epoch) = {1000 * train_loss:.3f}, '
              f'te_loss (epoch) = {1000 * test_loss:.3f}')
```
